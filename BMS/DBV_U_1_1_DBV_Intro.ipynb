# Teil 1: Einführung in scikit-image 

```python
# Laden der erforderlichen Pakete
from skimage import data, io, filters
import numpy as np
import matplotlib.pyplot as plt
```

## Bilder laden, darstellen und inspizieren

Hinweis: in dieser Übung verwenden wir ausschliesslich mitgelieferte Bilder, die Teil der scikit-image-Distribution sind. Eine Übersicht über alle mitgelieferten Bilder finden Sie unter dem folgenden Link:
https://scikit-image.org/docs/stable/auto_examples/data/plot_general.html#sphx-glr-auto-examples-data-plot-general-py

### Bild 1: Schwarzweissbild

```python


# Erstes Schwarzweissbild laden
# Beispielbild: camera
bild1 = data.camera()
```

```python
# Zeichnen des Bilds in die aktuelle Figure (Darstellung)
io.imshow(bild1) 

# Darstellen der Figure (nachdem alle darzustellenden Inhalte erzeugt sind)
io.show()
```

```python
# Speichertypus ausgeben (=> Hinweis: scikit-image speichert Bilder als NumPy ndarrays)
# Hinweis: Verwendung der Funktion 'type()'
type(bild1)

```

```python
# Bilddimensionen und -grösse ausgeben
# mit Hilfe der NumPy Array-Methoden .shape und .size
print(bild1.shape)
print(bild1.size)


```

```python
# Minimal-, Maximal- und Mittelwert ausgeben
# mit Hilfe der NumPy Array Methoden .min, .max, .mean

print(np.min(bild1),np.max(bild1),np.mean(bild1))


```

### Bild 2: Farbbild

```python
# RGB-Bild laden
# Beispielbild: astronaut
bild2=data.astronaut()


```

```python
# Bild darstellen
io.imshow(bild2)
io.show()


```

```python
# Bilddimensionen und -grösse ausgeben => was hat sich geändert?

print(bild2.shape)

```

## Bildinhalte von Bild 2 auswerten und verändern / Zugriff auf Pixelwerte

```python
# Pixelwert der 20. Zeile und 40. Kolonne ausgeben

bild2[20,40]
```

```python
# Pixelwert des Bilds 2 verändern, bspw. Pixel and Position 100, 100
# auf "schwarz" setzen

# Hinweis : korrekte Lösung => mehrkanalige Zuweisung ... = [ , , ]
# einkanalige Lösung funktioniert 

# Hinweis: das funktioniert auch (am Beispiel von "weiss") 


# Bild darstellen

bild2[100,100]=[255,0,0]

io.imshow(bild2)
io.show()
```

```python
# Slicing: erste 100 Zeile auf "grün" und Spalten > 400 auf "rot" setzen
# Tipp: Adressierung von Zeilen [von:bis], Adressierung von Spalten [,von:bis]


# Bild darstellen

bild2[0:512]=255,0,0
bild2[75:425,200:300]=255,255,255
bild2[200:300,75:425]=255,255,255


io.imshow(bild2)
io.show()
```

```python
# Bildausschnitt extrahieren
# Ecke links oben: Zeile 100, Spalte 100
# Ecke rechts unten: Zeile 400, Spalte 400



# Bild darstellen
bild3=data.astronaut()
io.imshow(bild3)
io.show()
ausschnitt=bild3[100:400,100:400]

io.imshow(ausschnitt)
io.show()

```

```python
# Einzelnen Kanal aus dem vorigen RGB-Bild extrahieren (z.B. "blau")

blau=ausschnitt[:,:,2]
io.imshow(blau)


# Bild darstellen

```
