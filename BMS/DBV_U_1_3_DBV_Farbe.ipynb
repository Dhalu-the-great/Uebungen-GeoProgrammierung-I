## Bilder subtrahieren und vergleichen

```python
# Graustufenbilder subtrahieren und vergleichen
# Subtrahieren Sie die beiden Graustufenbilder a) aus dem RGB-Bild
# und b) aus dem HSV-Bild und analysieren Sie die Grauwertdifferenzen
# (minimale, maximale, mittlere Differenz)

gray_diff = bild2 - hsv_v
print('gray_diff Min: ', gray_diff.min())
print('gray_diff Max: ', gray_diff.max())
print('gray_diff Mean:', gray_diff.mean())

# Stellen Sie die beiden Graustufenbilder sowie das Differenzbild
# in einem 1 x 3 Plot mit den 3 beschrifteten Bildern dar
fig,axs=plt.subplots(3,1)
axs[0].imshow(bild2)
axs[0].set_title('Gray')
axs[1].imshow(hsv_v)
axs[1].set_title("HSV")
axs[2].imshow(gray_diff)
axs[2].set_title("differenz")
fig.tight_layout()
plt.show()

```
