# Teil 2: Bilder lesen und schreiben

```python
# Laden der erforderlichen Pakete
from skimage import data, io, color, filters
import numpy as np
import matplotlib.pyplot as plt

# Importieren der Python-Skripte mit den Qt-Dialogen zum Auswählen eines Files
# Wichtig: die beiden Python-Skripte müssen sich im selben Verzeichnis
# wie dieses Jupyter Notebook befinden
import myopenfiledialog
import mysavefiledialog
```

## Bilddatei öffnen

```python
# Datei selektieren mittels Open File Dialog

# Lösung mit Qt Dialog - kopiert von / credits to:
# https://stackoverflow.com/questions/20790926/ipython-notebook-open-select-file-with-gui-qt-dialog

# File Open Dialog liefert den vollen Pfadnamen der Datei als Byte Object
input_file = myopenfiledialog.gui_fname()

#Umwandlung des Pfadnamens in einen utf8-String
input_file = input_file.decode('utf8')
```

```python
# Bilddatei öffnen und Bild laden
bild1 = io.imread(input_file)

# Zeichne das Bild in die aktuelle Figure (Darstellung)
io.imshow(bild1) 

# Darstellen der Figure (nachdem alle darzustellenden Inhalte erzeugt sind)
io.show() 
```

## Bild verändern

```python
# Bildumwandlung 1: was geschieht hier?
bild2 = color.rgb2hsv(bild1)

io.imshow(bild2)
io.show()
```

```python
# Bildumwandlung 2: was geschieht hier?
bild3 = bild2[:,:,2]

io.imshow(bild3)
io.show
```

## Bild abspeichern

```python
# Bilddatei mit neuem Qt Dialog speichern

# Vollständigen Datei- und Pfadnamen des zu speicherneden Files
# mittels graphischem Dialog bestimmen

# WICHTIG 1: Ergänzen Sie den Filenamen durch die gültige Fileendung
# eines verbreiteten Bildformats

# WICHTIG 2: bei gewissen Kombinationen von Bildinhalt (schwarzweiss) und Bildformat
# kann ein Warnhinweis erscheinen. Ignorieren Sie diesen.

output_file = mysavefiledialog.gui_fsname()

# Umwandlung des Pfadnamens in einen utf8-String (um Fehlermeldungen zu vermeiden)
output_file = output_file.decode('utf8')

io.imsave(output_file, bild3)
```
